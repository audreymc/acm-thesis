## Stationarity Tests

"[A key assumption in extreme value analysis is the stationarity of the data, which implies that the mean and standard deviation remain constant over time.](https://www.sciencedirect.com/science/article/pii/S2950601824000265#:~:text=A%20key%20assumption%20in%20extreme,deviation%20remain%20constant%20over%20time.)" Hence, in order to effectively apply EVT, we must test for stationarity of our price difference data. We can do so utilizing some common stationarity tests for time series data.

**See**: 
- [Detecting stationarity in time series data](https://medium.com/towards-data-science/detecting-stationarity-in-time-series-data-d29e0a21e638)

### 1. The Dickey-Fuller Test

Tests for presence of unit root within time series data.

**Python functions**
- ``statsmodels.tsa.stattools.adfuller``

### 2. KPSS Test

Tests for trend stationarity within time series data.

**Python functions**
- ``statsmodels.tsa.stattools.kpss``

## Kullback-Leibler (KL) divergence 

Quantifying the distance between distributions. Can be utilized for detecting distribution shift or assessing the fits of extreme value distributions.

**Python functions**
